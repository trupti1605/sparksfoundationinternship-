# Importing important libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# render the plot inline, instead of in a separate window
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing dataset 

```python
df = pd.read_csv("https://raw.githubusercontent.com/trupti1605/GitHub-example/master/SampleSuperstore.csv")
```

# Scatter Plot

```python
plotData = df[["Category","Profit","Quantity","Sales","Discount"]]
sns.set(style="ticks")
sns.pairplot(plotData, hue="Category")
```
